```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from urllib.request import urlopen
import requests
from bs4 import BeautifulSoup
```

```python
dfsectores=pd.read_excel(".xlsx")
```

```python
df=pd.read_excel("1_comex_sectores.xlsx")
```

```python
valor_a_multiplicar = (-1)
df['Importaciones'] = df['Importaciones'] * valor_a_multiplicar
df1=df.head(23)
df2=df.tail(46)
df3=df2.head(23)
df4=df2.tail(23)
```

```python
min_value = min(df1[['Importaciones', 'Exportaciones']].min())
max_value = max(df1[['Importaciones', 'Exportaciones']].max())

# Configurar el grfico
fig, ax = plt.subplots(figsize=(10, 6))

# Crear barras para las exportaciones
ax.bar(df1['Sectores Econmicos'], df1['Exportaciones'], color='red', alpha=0.5, label='Exportaciones')

# Crear barras para las importaciones, con valor cero en el centro del eje Y
importaciones_negativas = df1['Importaciones'].apply(lambda x: min(0, x))
importaciones_positivas = df1['Importaciones'].apply(lambda x: max(0, x))
ax.bar(df1['Sectores Econmicos'], importaciones_positivas, color='blue', alpha=0.5, label='Importaciones')
ax.bar(df1['Sectores Econmicos'], importaciones_negativas, color='blue', alpha=0.5)

# Crear lnea para la balanza comercial
balanza_comercial = df1['Exportaciones'] + df1['Importaciones']
ax.plot(df1['Sectores Econmicos'], balanza_comercial, color='green', marker='o', label='Balanza Comercial')

# Ajustar lmites del eje Y
plt.ylim(-max_value * 1.1, max_value * 1.1)

# Agregar etiquetas y leyenda
ax.set_title('Importaciones, Exportaciones y Balanza Comercial por Sector Econmico')
ax.set_xlabel('Sectores Econmicos')
ax.set_ylabel('Valor')
ax.axhline(0, color='black', linewidth=0.5)  # Lnea en cero
ax.legend()

# Rotar etiquetas del eje X
plt.xticks(rotation=45, ha='right')

# Mostrar el grfico
plt.tight_layout()
plt.show()
```

Por los datos ya presentados vemos que la mayor carencia que tiene Espaa se centra en la importacin de recursos que se transforman en energa (ya sea gasolina y otros combustibles como elecctricidad). Por ello voy a centrarme en analizar cmo se organiza la produccin energtica en Espaa y si es posible la evolucin del consumo de combustibles fsiles en Espaa de cara a la futura implantacin de coches elctricos.


Analicemos cmo ha ido evolucionando el sistema elctrico en Espaa a travs de su consumo y produccin

El siguiente grfico se usar para ver nicamente cuanto se produce y se consume, pero lo interesante de aqu es saber el porcentaje de cada una de las fuentes de energa.

```python
rel = requests.get('https://datosmacro.expansion.com/energia-y-medio-ambiente/electricidad-consumo/espana')
htmlel = rel.content
soupel = BeautifulSoup(htmlel, 'html.parser')
soupel.prettify
soupel.find_all('table', attrs={'class': 'table tabledat table-striped table-condensed table-hover'})
html_tableel = soupel.find_all('table', attrs={'class': 'table tabledat table-striped table-condensed table-hover'})[0]
dfelectricidad = pd.read_html(html_tableel.prettify())[0]
```

```python
y1 = dfelectricidad['Generacin GWh']
y2 = dfelectricidad['Consumo GWh'] 
x=dfelectricidad['Fecha']
plt.plot(x,y1,marker='d',linestyle='--', color='black', label='Generacin GWh')
plt.legend()
plt.xlabel('Fecha')
plt.ylabel('GWh')
plt.title('Relacin entre Produccin y Consumo de energa')
plt.bar(x,y2, label='Consumo GWh')
plt.figure(figsize=(12, 8))
plt.show()
```
